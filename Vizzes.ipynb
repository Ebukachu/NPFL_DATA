```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\USER\\Documents\\Football Analytics\\IKD_REMO_B.csv')
df
```

```python
#fill null values
df = df.fillna(value='null')
df
```

```python
#remove duplicates
df = df.drop_duplicates()
df
```

```python
# data cleaning
df['Action'] = df['Action'].replace({
    'Successful Shot Pass': 'Successful Short Pass',
    'Unsuccessful Shot Pass': 'Unsuccessful Short Pass'
}, regex=True)

df
```

```python
# Reset the DataFrame index to make it sequential
df = df.reset_index(drop=True)

# Plotting code
fig, ax = plt.subplots(figsize=(13.5, 8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

# Create the pitch
pitch = Pitch(pitch_type='opta', orientation='horizontal',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

# Draw the pitch on the ax figure
pitch.draw(ax=ax)

# Use a for loop to plot each pass
for i in range(len(df['x'])):
    if df['Action'][i] == 'Successful Progressive Pass':
        plt.plot((df['x'][i], df['xEnd'][i]), (df['y'][i], df['yEnd'][i]), color='green')
        plt.scatter(df['x'][i], df['y'][i], color='green')
    elif df['Action'][i] == 'Unsuccessful Progressive Pass':
        plt.plot((df['x'][i], df['xEnd'][i]), (df['y'][i], df['yEnd'][i]), color='red')
        plt.scatter(df['x'][i], df['y'][i], color='red')

plt.title('Ikorodu City vs Remo Stars B Progressive Passes\nSuccessful Progressive Passes: Green | Unsuccessful Progressive Passes: Red', 
          color='white', size=20)
plt.show()
```

```python
shots = ['Shot off target', 'Shot on target']

new_df = df[df['Action'].isin(shots)]
new_df
```

```python
text_color = 'white'  # Define your desired color here


fig, ax = plt.subplots(figsize=(13,8.5))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

pitch = Pitch(pitch_type='opta', orientation='vertical',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True, view='full')

pitch.draw(ax=ax)


# Filter data for 'Shot on target' and 'Shot off target'
shots_on_target = new_df[new_df['Action'] == 'Shot on target']
shots_off_target = new_df[new_df['Action'] == 'Shot off target']

# Plot 'Shot on target' as green points
plt.scatter(shots_on_target['x'], shots_on_target['y'], s=100, c='green', alpha=.7, label='Shot on target')

# Plot 'Shot off target' as red points
plt.scatter(shots_off_target['x'], shots_off_target['y'], s=100, c='red', alpha=.7, label='Shot off target')


# Add text to the figure using plt.text
plt.title('Ikorodu City vs Remo Stars B Shot Chart\nShot on Target: Green | Shot off Target: Red', 
          color='white', size=20)

# Calculate total shots
total_shots = len(new_df)
plt.text(x=0.27, y=0.67, s=f'Total Shots: {total_shots}', fontsize=14, fontfamily='Sans-serif', color=text_color, transform=ax.transAxes)

plt.show()
```
